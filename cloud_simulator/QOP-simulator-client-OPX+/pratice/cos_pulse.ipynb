```python
t_total = 200
t_edge = t_total/10
#t_raise, t_down, t_flat = 10,10,800  # ns in unit
Amplitude=2

def raise_cosine(t,T):
    return 0.5*(1-np.cos(2*np.pi*t/T))
```

```python
#t = np.linspace(0,t_raise+t_down,t_raise+t_down) # ns in unit
t =  np.arange(0,t_edge,1) # ns in unit
raise_down_part = raise_cosine(t,t_edge)
#plt.plot(t,raise_down_part)

#t_square = np.linspace(0,t_raise+t_down+t_flat,t_raise+t_down+t_flat) # ns in unit
t_square = np.arange(0,t_total+t_edge,1) # ns in unit
flat = np.ones(t_total)
square = np.concatenate((raise_down_part[0:len(raise_down_part)//2],flat,raise_down_part[len(raise_down_part)//2:]),axis=0)
plt.plot(t_square,square*Amplitude)

plt.show()
```

```python
def raise_cosine(t,T):
    return 0.5*(1-np.cos(2*np.pi*t/T))

def rasie_cosine_puilse(duration:int):
    #t_edge = int(duration/10)
    t_edge = int(duration/10)
    t_edge = 20
    t =  np.arange(0,t_edge+1,1) # ns in unit
    raise_down_part = raise_cosine(t,t_edge)
    flat = np.ones(duration-t_edge)
    square = np.concatenate((raise_down_part[0:len(raise_down_part)//2],flat,raise_down_part[len(raise_down_part)//2+1:]),axis=0)
    return square
```

```python
plt.plot(rasie_cosine_puilse(200))
```

```python
for i in range(10):
    plt.plot(rasie_cosine_puilse(20+i*10))
plt.show()
```
